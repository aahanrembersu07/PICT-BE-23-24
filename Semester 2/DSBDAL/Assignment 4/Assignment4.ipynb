```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("titanic.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
df.isnull().sum()
```

```python
df['Cabin']=df['Cabin'].fillna(df['Cabin'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum()
```

```python
sns.boxplot(x='Age',y='Survived',data=df)
```

```python
sns.histplot(x='Age',data=df,kde=True)
```

```python
df.describe()
```

```python
bins=[0,7.9,14.45,31,512.33]
labels=['1st','2nd','3rd','4th']
df['Fare Category']=pd.cut(df['Fare'],bins=bins,labels=labels)
x=sns.countplot(x="Fare Category",hue="Survived",data=df).legend(labels=["Deceased","Survived"])
x.set_title("Survival based on Fare Category")
```

```python
df['Fare'].value_counts()
```

```python
Q1=df['Fare'].quantile(0.25)
Q3=df['Fare'].quantile(0.75)
IQR=1.5*(Q3-Q1)
lb=Q1-IQR
ub=Q3+IQR
outliers=df[(df['Fare']<lb)|(df['Fare']>ub)]
outliers
```

```python
filtered_df=df[(df['Fare']>=lb) & (df['Fare']<=ub)]
sns.histplot(x="Fare",hue="Survived",data=filtered_df,kde=True)
```

```python
sns.boxplot(x="Sex",y="Age",hue="Survived",data=filtered_df)
```
