```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("titanic.csv")
df.head()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
df.isnull().sum()
```

```python
df['Cabin']=df['Cabin'].fillna(df['Cabin'].mode()[0])
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
df.isnull().sum()
```

```python
sns.boxplot(x="Sex",y="Age",hue="Survived",data=df)
```

```python
survivors=df.groupby(['Sex','Survived']).size().unstack()
survivors.columns=['Non-Survivors','Survivors']
survivors
```

```python
df.describe()
```

```python
Q1=df['Age'].quantile(0.25)
Q3=df['Age'].quantile(0.75)
IQR=1.5*(Q3-Q1)
lb=Q1-IQR
ub=Q3+IQR
print(lb)
print(ub)
outliers=df[(df['Age']<lb) | (df['Age']>ub)]
outliers
```

```python
filtered_df=df[(df['Age']>=lb) & (df['Age']<=ub)]
```

```python
sns.boxplot(x="Sex",y="Age",hue="Survived",data=filtered_df)
```
