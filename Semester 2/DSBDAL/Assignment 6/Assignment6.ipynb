```python
def feature(df):
    numerics=df.select_dtypes(include=['number']).columns.tolist()
    nominal=df.select_dtypes(exclude=['number']).columns.tolist()
    return (numerics,nominal)
feature(df)
```

```python
Q1=df['SepalLengthCm'].quantile(0.25)
Q3=df['SepalLengthCm'].quantile(0.75)
IQR=1.5*(Q3-Q1)
lb=Q1-IQR
ub=Q3+IQR
print(lb,ub)
filtered_df=df[(df['SepalLengthCm']>=lb) & (df['SepalLengthCm']<=ub)]
sns.histplot(x='SepalLengthCm',hue='Species',data=filtered_df,kde=True)
```
